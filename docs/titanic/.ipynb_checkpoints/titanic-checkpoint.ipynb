# EDA Danych dotyczących Titanica

## O Danych

Dane o pasażerach Titanica

Zbiór danych zawiera informacje o pasażerach RMS Titanic, który zatonął 15 kwietnia 1912 roku po zderzeniu z górą
lodową. Dane obejmują takie atrybuty jak klasa podróży, wiek, płeć, liczba rodzeństwa/małżonków na pokładzie,
liczba rodziców/dzieci na pokładzie, cena biletu oraz miejsce zaokrętowania.

Zbiór zawiera także informację o tym, czy pasażer przeżył katastrofę.

Titanic przewoził ponad 2,200 osób, z czego ponad 1,500 zginęło, co czyni tę katastrofę jedną z najbardziej
tragicznych w historii morskiej.

Kolumny:

* **pclass** - Klasa biletu
* **survived** - Czy pasażer przeżył katastrofę
* **name** - Imię i nazwisko pasażera
* **sex** - Płeć pasażera
* **age** - Wiek pasażera
* **sibsp** - Liczba rodzeństwa/małżonków na pokładzie
* **parch** - Liczba rodziców/dzieci na pokładzie
* **ticket** - Numer biletu
* **fare** - Cena biletu
* **cabin** - Numer kabiny
* **embarked** - Port, w którym pasażer wszedł na pokład (C = Cherbourg, Q = Queenstown, S = Southampton)
* **boat** - Numer łodzi ratunkowej
* **body** - Numer ciała (jeśli pasażer nie przeżył i ciało zostało odnalezione)
* **home.dest** - Miejsce docelowe

### funkcja wyświetlająca charakterystyki kolumn

```python
def df_summary(df: pd.DataFrame) -> pd.DataFrame:
    """
    Tabela podsumowania kolumn DataFrame:
    - dtype: typ danych
    - non_null: liczba niepustych wartości
    - is_null: liczba pustych wartości
    - unique: liczba unikalnych wartości
    - duplicates: liczba duplikatów (non_null - unique)
    - is_null_pct: procent wartości pustych względem liczby wierszy
    - outliers: liczba wartości odstających wg metody IQR (dla kolumn numerycznych)
    - outliers_pct: procent wartości odstających względem liczby niepustych
    """
    nunique   = df.nunique(dropna=True)
    dtypes    = df.dtypes
    non_null  = df.notna().sum()
    is_null   = df.isna().sum()
    rows      = len(df)

    duplicates = non_null - nunique
    is_null_pct = (is_null / rows * 100).round(2)

    # Oblicz outliery tylko dla kolumn numerycznych
    outliers = []
    outliers_pct = []

    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            s = df[col].dropna()
            if len(s) > 0:
                Q1 = s.quantile(0.25)
                Q3 = s.quantile(0.75)
                IQR = Q3 - Q1
                lower = Q1 - 1.5 * IQR
                upper = Q3 + 1.5 * IQR
                n_out = ((s < lower) | (s > upper)).sum()
                pct_out = (n_out / len(s) * 100).round(2)
                outliers.append(int(n_out))
                outliers_pct.append(pct_out)
            else:
                outliers.append(np.nan)
                outliers_pct.append(np.nan)
        else:
            outliers.append(np.nan)
            outliers_pct.append(np.nan)

    summary = pd.DataFrame({
        "dtype": dtypes,
        "non_null": non_null,
        "is_null": is_null,
        "unique": nunique,
        "duplicates": duplicates,
        "is_null_pct": is_null_pct,
        "outliers": outliers,
        "outliers_pct": outliers_pct
    })

    summary.reset_index(inplace=True)
    summary.rename(columns={"index": "column"}, inplace=True)
    summary.index.name = "col_id"
    return summary
```

```python
df_summary(df)
```

### % pasażerów, któzy przeżyli katastrofę w podziale na płeć

```python
df_surv_sex = df.groupby(['sex', 'survived']).size().to_frame('count')
df_surv_sex["pct"]=(df_surv_sex["count"] / df_surv_sex.groupby(level=0)["count"].transform("sum") * 100).round(2)
df_surv_sex
```

```python
# spłaszczamy index
df_plot = df_surv_sex.reset_index()

plt.figure(figsize=(6,5))

# rysujemy stacked bar
bottoms = {sex:0 for sex in df_plot["sex"].unique()}
colors = {0:"tab:red", 1:"tab:green"}

for survived_val in [0,1]:
    subset = df_plot[df_plot["survived"]==survived_val]
    plt.bar(
        subset["sex"],
        subset["pct"],
        bottom=[bottoms[sex] for sex in subset["sex"]],
        label=f"survived={survived_val}",
        color=colors[survived_val]
    )
    # aktualizujemy "bottom", żeby kolejne wartości nakładały się
    for sex, pct in zip(subset["sex"], subset["pct"]):
        bottoms[sex] += pct

plt.ylabel("Procent [%]")
plt.title("Przeżywalność w zależności od płci")

# legenda poza wykresem, po prawej
plt.legend(title="Survived", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()
plt.show()
```

- przeżyło 75% kobiet, a tylko 19% mężczyzn - bardzo duża różnica. Prawdopodobnie mężczyźni musieli ratowali kobiety z narażeniem własnego życia

- ponad połowa pasażerów wybrała 3 czyli najtańszą klasę biletów
- ilość pasażerów, któa wybrała klasę 1 i 2 porównywalna

### Rozkład klasy biletu względem pasażerów i przeżywalności

```python
df_surv_pclass = df.groupby(['pclass', 'survived']).size().to_frame('count')
df_surv_pclass["pct"]=(df_surv_pclass["count"] / df_surv_pclass.groupby(level=0)["count"].transform("sum") * 100).round(2)
df_surv_pclass
```

```python
## spłaszczamy indeks
df_plot = df_surv_pclass.reset_index()

plt.figure(figsize=(7,5))
ax = plt.gca()

# kolory: zielony dół, różowy góra
colors = {1: "mediumseagreen", 0: "lightcoral"}

bottoms = {p:0 for p in df_plot["pclass"].unique()}

# najpierw survived=1 (zielony dół)
for surv_val in [1,0]:
    subset = df_plot[df_plot["survived"]==surv_val]
    sns.barplot(
        data=subset,
        x="pclass", y="pct",
        color=colors[surv_val],
        ax=ax,
        label=f"survived={surv_val}",
        bottom=[bottoms[p] for p in subset["pclass"]]
    )
    # aktualizacja wysokości "bottom"
    for p, pct in zip(subset["pclass"], subset["pct"]):
        bottoms[p] += pct

ax.set_ylabel("Procent [%]")
ax.set_xlabel("Klasa biletu (pclass)")
ax.set_title("Przeżywalność według klasy biletu")
ax.legend(title="Survived", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

- przeżywalność w 3 klasie 25% vs 62% w klasie 1. Wyraźnie widać, że bogaci pasażerowie musieli mieć pierwszeństwo w dostępie do łodzi ratunkowych i można wysnuć przypuszczenie, że byli mniej zainteresowani ratowaniem kobiet.

### Sprawdzenie przeżywalności mężczyzn w zależności od klasy biletu

```python
# grupowanie po klasie, płci i przeżyciu
df_surv_pclass_sex = (
    df.groupby(['pclass', 'sex', 'survived'])
      .size()
      .to_frame('count')
)

# procenty w obrębie każdej grupy pclass+sex
df_surv_pclass_sex["pct"] = (
    df_surv_pclass_sex["count"] /
    df_surv_pclass_sex.groupby(level=[0, 1])["count"].transform("sum") * 100
).round(2)

# odfiltrowanie tylko tych, którzy przeżyli (survived=1)
df_surv_pclass_sex = df_surv_pclass_sex.xs(1, level="survived")
df_surv_pclass_sex
```

- tutaj przeżywalność mężczyzn w 1 i 2 klasie spada do około 15%, gdzie w 1 wynosi 34%
- ciekawe jest to, że przeżywalność kobiet w 1 i 2 klasie była bardzo duża - odpowiednio 96% i 88%

### Rozkład pasażerów na łodziach ratunkowych w podziale na płeć

```python
import numpy as np

# 0 jeśli boat == "unknown", inaczej 1
df["boat01"] = np.where(df["boat"] == "unknown", 0, 1)

# grupowanie po boat01 i sex
df_boats = df.groupby(["boat01", "sex"]).size().to_frame("count")

# policz procent w obrębie każdej płci
df_boats["pct"] = (
    df_boats["count"] /
    df_boats.groupby(level=0)["count"].transform("sum") * 100
).round(2)

df_boats
```

- na łodziach było 34% mężczyzn

### Rozkład pasażerów w podziale na klasę biletu, płeć i miejsce w łodzi ratunkowej

```python
import numpy as np

# 0 jeśli boat == "unknown", inaczej 1
df["boat01"] = np.where(df["boat"] == "unknown", 0, 1)

# grupowanie po boat01, pclass i sex
df_boats = df.groupby(["boat01", "pclass", "sex"]).size().to_frame("count")

# policz procent w obrębie każdej grupy boat01+pclass
df_boats["pct"] = (
    df_boats["count"] /
    df_boats.groupby(level=[0, 1])["count"].transform("sum") * 100
).round(2)

df_boats
```

```python
# spłaszczamy index, żeby mieć normalne kolumny
df_plot = df_boats.reset_index()

plt.figure(figsize=(10,6))
g = sns.catplot(
    data=df_plot,
    x="pclass", y="pct",
    hue="sex",
    col="boat01",      # osobny panel dla boat01=0 i boat01=1
    kind="bar",
    height=5, aspect=1
)

g.set_axis_labels("Klasa biletu (pclass)", "Procent [%]")
g.set_titles("boat01 = {col_name}")
plt.show()
```

- tutaj wyraźnie widać jak dużą liczbę kobiet udało się uratować i to niezależnie od klasy biletu
- ciekawym jest, że dużo mężczyzn z 3 klasą biletu znalazło się na łodziach
- przypuszczenie: im nizsza klasa tym nizsze poczucie odpowiedzialności i skłonnosci do poswieceń u mężczyzn.
